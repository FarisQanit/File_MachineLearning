```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

tabel_data = pd.read_csv("Pumpkin_Seeds_Dataset.csv")
tabel_data.hist(column = 'Roundness',bins=30)
```

```python
tabel_data[["Solidity","Roundness"]].corr()
```

```python
import seaborn as sns
sns.regplot(x="Solidity",y="Roundness",data=tabel_data)
```

```python
tabel_data[["Equiv_Diameter","Extent"]].corr()
```

```python
sns.regplot(x="Equiv_Diameter",y="Extent",data=tabel_data)
```

```python
tabel_data[["Aspect_Ration","Compactness"]].corr()
```

```python
sns.regplot(x="Aspect_Ration",y="Compactness",data=tabel_data)
```

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

tabel_data['Aspect_Ration'].plot.line(ax=ax[0][0])
ax[0][0].set_title('Aspect_Ration')

tabel_data['Compactness'].plot.line(ax=ax[0][1])
ax[0][1].set_title('Compactness')

tabel_data['Equiv_Diameter'].plot.line(ax=ax[1][0])
ax[1][0].set_title('Equiv_Diameter')

tabel_data['Extent'].plot.line(ax=ax[1][1])
ax[1][1].set_title('Extent')
```
